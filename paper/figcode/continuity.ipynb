```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2*np.pi, 50)
```

```python
X, Y = np.meshgrid(x,x)
```

```python
C = np.sin(X)*np.cos(Y)
```

```python
fig, axd  = plt.subplot_mosaic([['table', 'lines'],['table', 'image']], constrained_layout=True)
tab = axd['table'].table(C[::5, ::5], loc='center')
axd['table'].set_title("data", y=.8)
axd['table'].axis('off')
axd['lines'].plot(C)
axd['lines'].set_title("plt.plot(data)")
axd['lines'].set(xlim=(0, len(x)-1))
axd['image'].imshow(C)
axd['image'].set_title("plt.imshow(data)")
fig.savefig("../figures/continuity.png")
```
